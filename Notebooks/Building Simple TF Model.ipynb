```python
df = pd.read_csv('/kaggle/input/devanagari-character-set/data.csv')

X = df.iloc[:,:-1]
X_images = X.values.reshape(X.shape[0], 32, 32)
y = df.iloc[:,-1]
```

```python
# output in binary format
from sklearn.preprocessing import LabelBinarizer
binencoder = LabelBinarizer()
Y = binencoder.fit_transform(y)
```

```python
X = X / 255
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.15, random_state=786)
```

```python
# Building Simple TF Model
from keras.models import Sequential
from keras.layers import Dense
from keras.models import Model
```

```python
model = Sequential()

# First Dense layer for All Inputs - 92K pixel values
model.add(Dense(
    1024, 
    input_dim=1024, 
    kernel_initializer='normal', 
    activation='relu'
))

# Output Layer with Number of Output Classes
model.add(Dense(
    46, 
    kernel_initializer='normal', 
    activation='softmax'
))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
epochs = 7
batch_size = 350
history = model.fit(X_train.values, y_train, validation_split=0.15, epochs=epochs, batch_size=batch_size, verbose=2)
```

```python
scores = model.evaluate(X_test.values, y_test, verbose=0)
print("Accuracy: %.2f%%" % (scores[1]*100))
```
